```python
#Loading necesssary libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#Loading IPL Matches dataset
ipl=pd.read_csv("C:/Users/ASUS/Downloads/matches.csv")
```

```python
#Having a glance at first five rows of the dataset
ipl.head()
```

```python
#Analyzing the no of rows and columns
ipl.shape
```

```python
#Finding the player with most 'Player of the Match' award
ipl['player_of_match'].value_counts()
```

```python
#Finding the top 10 players with most number of 'Player of the Match' awards
ipl['player_of_match'].value_counts()[0:10]
```

```python
#Finding the top 5 players with most number of 'Player of the Match' awards
ipl['player_of_match'].value_counts()[0:5]
```

```python
#Making a list of the top 5 players with most number of 'Player of the Match' awards
list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
#Creating a barplot of the top 5 players with most number of 'Player of the Match' awards
plt.figure(figsize=(8,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color='purple')
plt.show()
```

```python
#Creating a barplot of the top 10 players with most number of 'Player of the Match' awards
plt.figure(figsize=(16,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:10].keys()),list(ipl['player_of_match'].value_counts()[0:10]),color='yellow')
plt.show()
```

```python
#Getting the frequency of 'result' column
ipl['result'].value_counts()
```

```python
#Finding the number of toss wins with respect to each team
ipl['toss_winner'].value_counts()
```

```python
#Extracting the details of team which won after batting first
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
#Looking into the first five data
batting_first.head()
```

```python
#Plotting a histogram for understanding the distribution of runs
plt.figure(figsize=(5,7))
plt.hist(batting_first['win_by_runs'])
plt.title('Distribution of Runs')
plt.xlabel('Runs')
plt.show()
```

```python
#Finding the number of wins by each team after batting first
batting_first['winner'].value_counts()
```

```python
#PLotiing a barplot with top 3 teams with most wins after batting first
plt.figure(figsize=(6,6))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=["blue","yellow","red"])
plt.show()        
```

```python
#PLotiing a piechart with teams having most wins after batting first
plt.figure(figsize=(15,15))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()        
```

```python
#Extracting the details of team which won after batting second
batting_second=ipl[ipl['win_by_wickets']!=0]
```

```python
#Looking into the first five data
batting_second.head()
```

```python
#Plotting a histogram for understanding the wins based on wickets
plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.show()
```

```python
#Finding the number of wins by each team after batting second
batting_second['winner'].value_counts()
```

```python
#PLotiing a barplot with top 3 teams with most wins after batting second
plt.figure(figsize=(7,6))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color=["green","pink","grey"])
plt.show()        
```

```python
#PLotiing a piechart with teams having most wins after batting second
plt.figure(figsize=(15,15))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()        
```
